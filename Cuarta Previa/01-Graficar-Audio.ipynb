```python
# GRAFICAR ARCHIVO DE AUDIO

# Importar las librerías 
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
#este programa lo que hace es que utiliza una librería llamada wavfile tomado de scipy.io y lo que hace es básicamente 
#es permitir por ejemplo leer un archivo de audio con extensión wap.

# Lee el archivo de audio
frecuencia_muestreo, senial = wavfile.read('sonido_aleatorio.wav')
#entonces lo lee y cuando lo lee lo graba en una variable llamada senial, también graba y nos dice cual es la 
#frecuencia de ese archivo.

# Display the params
print('\nTamaño señal:', senial.shape)
print('Tipo de dato:', senial.dtype)
print('Duracción de la señal:', round(senial.shape[0] / float(frecuencia_muestreo), 2), 'seconds')
print('Frecuencia de muestreo: ', frecuencia_muestreo)
#una vez que lo traemos a la memoria entonces lo imprimimos con unas funciones. El tamaño de la señal, el tipo de datos, 
#la uracion de la señal y la frecuencia de muestreo, que es una variable que devuelve directamente el.


# Normalizar la señal
senial = senial / np.power(2, 15)
#luego se normaliza la señal, porque esta señal puede tener cualquier tipo o cualquier tamaño. Entonces se normaliza.

# Extraer los primeros 50 valores
senial = senial[:50]
#extraemos los primeros 50 valores

# Construir el eje de tiempo en milisegundos
eje_del_tiempo = 1000 * np.arange(0, len(senial), 1) / float(frecuencia_muestreo)
#calculamos el eje del tiempo para poderlo graficar 

# Dibujar la señal de audio
plt.plot(eje_del_tiempo, senial, color='black')
plt.xlabel('Tiempo (milisegundos)')
plt.ylabel('Amplitud')
plt.title('Señal Entrada de Audio')
plt.show()
#y luego lo dibujamos, y cuando lo hagamos aparece la señal de audio en el tiempo.
```
